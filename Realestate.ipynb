```python
create_agents_table = """
CREATE TABLE IF NOT EXISTS agents (
    agent_id VARCHAR(10) PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    phone VARCHAR(20),
    email VARCHAR(100),
    commission_rate DECIMAL(5,2),
    deals_closed INT,
    rating DECIMAL(3,2),
    experience_years INT,
    avg_closing_days DECIMAL(5,2)
)
"""

cursor.execute(create_agents_table)
conn.commit()

print("Agents table created successfully!")
```

```python
create_sales_table = """
CREATE TABLE IF NOT EXISTS sales (
    listing_id VARCHAR(10) PRIMARY KEY,
    sale_price DECIMAL(15,2),
    date_sold DATE,
    days_on_market INT,
    FOREIGN KEY (listing_id) REFERENCES listings(listing_id)
)
"""

cursor.execute(create_sales_table)
conn.commit()

print("Sales table created successfully!")
```

```python
create_buyers_table = """
CREATE TABLE IF NOT EXISTS buyers (
    buyer_id VARCHAR(10) PRIMARY KEY,
    sale_id VARCHAR(10),
    buyer_type VARCHAR(50),
    payment_mode VARCHAR(30),
    loan_taken BOOLEAN,
    loan_provider VARCHAR(100),
    loan_amount DECIMAL(15,2)
)
"""

cursor.execute(create_buyers_table)
conn.commit()

print("Buyers table created successfully!")
```

```python
import pandas as pd

buyers_df = pd.read_json(
    r"C:\Users\admin\Desktop\brickview\buyers_cleaned.json"
)

print(buyers_df.head())
print(buyers_df.columns.tolist())
```

```python
query = """
INSERT INTO buyers
(buyer_id, sale_id, buyer_type,
 payment_mode, loan_taken,
 loan_provider, loan_amount)

VALUES (%s,%s,%s,%s,%s,%s,%s)
"""
```

```python
print(buyers_df.head())
```

```python
data_list = []

for _, row in buyers_df.iterrows():
    data_list.append((
        str(row["buyer_id"]),
        str(row["sale_id"]),
        str(row["buyer_type"]),
        str(row["payment_mode"]),
        bool(row["loan_taken"]) if pd.notna(row["loan_taken"]) else None,
        None if pd.isna(row["loan_provider"]) else str(row["loan_provider"]),
        None if pd.isna(row["loan_amount"]) else float(row["loan_amount"])
    ))
cursor.executemany(query, data_list)
conn.commit()

print("buyers inserted successfully!")
```
